# Degree Distribution and other Properties

```python
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
from networkx.algorithms import approximation as appx
print("Nodes", G.number_of_nodes())
print("Edges", G.number_of_edges())
print("Avg. Clustering Coefficient", appx.average_clustering(G, trials=300))
print("Avg. Shortest Path", nx.average_shortest_path_length(G))
```

```python
# Degree Distributions

deg = dict(nx.degree(G))
deg_dist = Counter(deg.values())
# Splits Counter key-value pairs
x, y = zip(*deg_dist.items())
y = [val / G.number_of_nodes() for val in y]

plt.plot(x, y, marker='.', linestyle='None')
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Degree")
plt.ylabel("Probability")
plt.title("Degree Distribution (log-log scale)")
plt.show()

plt.plot(x, y, marker='.', linestyle='None')
plt.xscale("linear")
plt.yscale("linear")
plt.xlabel("Degree")
plt.ylabel("Probability")
plt.title("Degree Distribution (linear scale)")
plt.show()
```

```python
# CCDF Alpha
X = np.asarray(x)
X = np.log(X)
A = np.vstack([X,np.ones(len(x))]).T
Y = np.asarray(y)
# Does reverse cumulative sum to get CCDF values
Y[::-1].cumsum()[::-1]
Y = np.log(Y)
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print("Alpha is around: ",((m*-1)+1))

plt.scatter(x, y, label='empirical')

def ESTpdf(x):
    # Using m and c as derived above to get normalization constant and alpha value
    return (((-m)*np.exp(c)) * pow(x, -((m*-1)+1)))

ESTpdfX = np.arange(min(x), max(x))
ESTpdfY = ESTpdf(ESTpdfX)
plt.plot(ESTpdfX, ESTpdfY, label='estimated pdf')


plt.legend(loc='upper right')
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Degree")
plt.ylabel("Probability")
plt.title("Empirical & Estimated (using CCDF regression)")
plt.show()
```

```python
# Now working with Girvan Newman, since things before were based on starting
# graph and didn't adapt.
# Let's see if we see somthing new. Now, we are deleting those nodes though! Wild!
```

```python
from networkx.algorithms import centrality

def custom_girvan_newman(G, k, z=10):
    if (nx.number_connected_components(G) >= k):
        # Returns if network already split into clusters >= k
        return G
    
    H = G.copy()
    removed_nodes = []
    while (nx.number_connected_components(H) < k and H.number_of_edges() > 0 and len(removed_nodes) < z/2):
        # Removes "edge" with highest betweenness centrality before recomuting and repeats
        # Continues until reach desired k or run out of edges
        edge_centrality = centrality.edge_betweenness_centrality(H)
        top_edge = max(edge_centrality, key=edge_centrality.get)
        removed_nodes.append(top_edge)
        H.remove_nodes_from(top_edge)
    return H, removed_nodes

NewGraph, removed_nodes = custom_girvan_newman(G, 2)
print(removed_nodes)
```

```python
NewerGraph, newer_removed_nodes = custom_girvan_newman(NewGraph, 3)
print(newer_removed_nodes)
```
